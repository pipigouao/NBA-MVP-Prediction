# 数据清洗

```python
import numpy as np
import pandas as pd
```

## mvp球员数据清洗

```python
mvps = pd.read_csv("mvps.csv",header = 1)
```

```python
mvps.head()
```

```python
mvps.rename(columns={'Unnamed: 21':'Year'},inplace=True)
mvps.head()
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

## 球员数据清洗

```python
players = pd.read_csv("players.csv")
players.head()
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
```

```python
players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
players.index = players.index.droplevel()
players
```

## 连接球员数据和MVP投票数据

```python
## 去掉player前的*，便于后续连接
players["Player"] = players["Player"].str.replace("*","")
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
combined.head()
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined.head()
```

## 球队数据清洗

```python
teams = pd.read_csv("teams.csv")
teams.head()
```

```python
del teams["Unnamed: 0"]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
teams.head()
```
